무신사 페이지를 webdriver로 열기

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
url = 'https://www.musinsa.com/app/goods/1108007?utm_source=naver_jisicshopping&utm_medium=sh&source=NVSH&utm_term=%7Bkeyword%7D&utm_content=%7Bad_group%7D&NaPm=ct%3Dle2bdj00%7Cci%3D6965a3d3cef1aecca378eab34e393b9561d07f09%7Ctr%3Dslcc%7Csn%3D204973%7Chk%3D6f013c099daeba3f4886049540183c553f90681e'
driver = webdriver.Chrome('../driver/chromedriver')
driver.get(url)
```

웹페이지의 일반리뷰 버튼 클릭

```python
Review_nomal_button = driver.find_element(By.CSS_SELECTOR, '#estimate_goods')
Review_nomal_button.click()
```

타임슬립을 적용하지 않으면 시스템에서 기계로 인식후 오류 축력 --- time.sleep을 이용해 시간지연

```python
import time
```

```python
Reviews = []

for _ in range(1, 200):
    for page in range(4,8):
        for i in range(1,11):
            # 리뷰 전체의 바디를 가져오기
            Review_body = driver.find_element(By.CSS_SELECTOR, '#reviewListFragment') 
            # 리뷰의 한페이지에 10개의 리뷰가 존재하고 그 리뷰를 Review_list를 가져온다
            Review_list = Review_body.find_elements(By.CSS_SELECTOR, f'#reviewListFragment > div:nth-child({i}) > div.review-contents > div.review-contents__text')  
            # Review_list에서 review를 추출하고 Reviews라는 리스트에 더해준다.
            for idx, review in enumerate(Review_list):
                try:
                    Reviews.append(review.text)
                except:
                    print(review)
         
        Page_button = driver.find_element(By.CSS_SELECTOR, f'#reviewListFragment > div.nslist_bottom > div.pagination.textRight > div > a:nth-child({page})')
        Page_button.click()
        time.sleep(1)
        if page == 7:
            for i in range(1,11):
                Review_body = driver.find_element(By.CSS_SELECTOR, '#reviewListFragment')
                Review_list = Review_body.find_elements(By.CSS_SELECTOR, f'#reviewListFragment > div:nth-child({i}) > div.review-contents > div.review-contents__text')

                for idx, review in enumerate(Review_list):
                    try:
                        Reviews.append(review.text)
                    except:
                        print(review)
    Next_button = driver.find_element(By.CSS_SELECTOR, '#reviewListFragment > div.nslist_bottom > div.pagination.textRight > div > a.fa.fa-angle-right.paging-btn.btn.next')
    Next_button.click()
    time.sleep(1)
```
